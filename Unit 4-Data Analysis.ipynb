# Simple Linear Regression

```python
import numpy as np 
from sklearn.linear_model import LinearRegression
x=np.array([3,2,7,4,8]).reshape((-1,1))
y=np.array([6,1,8,5,9])
print(x)
print(y)
model=LinearRegression()
model.fit(x,y)
r_sq=model.score(x,y)
print(r_sq)
print(model.intercept_)
print(model.coef_)
score=model.predict(x)
print(score)
```

# Multiple Linear Regression

```python
import numpy as np 
from sklearn.linear_model import LinearRegression
x=np.array([[3,8],[4,5],[5,7],[6,3],[2,1]])
y=np.array([-3.7,3.5,2.5,11.5,5.7])
print(x)
print(y)
model=LinearRegression()
model.fit(x,y)
r_sq=model.score(x,y)
print(r_sq)
print(model.intercept_)
print(model.coef_)
score=model.predict([[3,2]])
print(score)
```

# MLR with 3 Independent Variables

```python
import numpy as np 
from sklearn.linear_model import LinearRegression
x=np.array([[3,8,2],[4,5,3],[5,7,4],[6,3,5],[2,1,6]])
y=np.array([-3.7,3.5,2.5,11.5,5.7])
print(x)
print(y)
model=LinearRegression()
model.fit(x,y)
r_sq=model.score(x,y)
print(r_sq)
print(model.intercept_)
print(model.coef_)
score=model.predict([[3,2,4]])
print(score)
```

# Polynomial Regression

```python
import numpy as np
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
x=np.array([[1,1],[2,4],[3,9],[4,16],[5,25]])
y=np.array([10,9,5,6,7])
poly=PolynomialFeatures(degree=2)
x_poly=poly.fit_transform(x)

model = LinearRegression()
model.fit(x_poly, y)
print(model.intercept_)
print(model.coef_)
prediction = model.predict(poly.transform([[3,2]]))
print("Prediction for [3, 2]:", prediction)
```

# Comparision of Linear and Polynomial Regression

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures


x=np.array([1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]).reshape(-1,1)
y=[100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

# Create Linear Regression model

lin=LinearRegression()
#fitting model using linear regression

lin.fit(x,y)
print("Linear intercept=",lin.intercept_)
print("Linear Coef=",lin.coef_)
print("Linear r score=",lin.score(x,y))

plt.scatter(x,y)
plt.plot(x,lin.predict(x))
plt.title("Linear Regression")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()

# Define PolynomialFeatures with degree 3
poly=PolynomialFeatures(degree=3)

# Fit the model using polynomial regression
x_poly=poly.fit_transform(x)
lin2=LinearRegression()
lin2.fit(x_poly,y)
print("\nPolynomial intercept=",lin2.intercept_)
print("Polynumial coef=",lin2.coef_)
print("Polynomial r score=",lin2.score(x_poly,y))

plt.scatter(x,y)
plt.plot(x,lin2.predict(x_poly),color='g')
plt.title("Polynomial Regression")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()

p=23
parray=np.array([[p]])
linear=lin.predict(parray)
print("Linear Prediction=",linear)

p=23
parray=np.array([[p]])
linear2=lin2.predict(poly.fit_transform(parray))
print("Polynomial Prediction=",linear2)
```

# Train Data and Test Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
x=np.array([1,2,3,5,6,8,10,15,20])
y=np.array([100,90,80,70,60,70,80,90,100])
plt.scatter(x,y,color='green')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
train_x=x[:7]
train_y=y[:7]
test_x=x[7:]
test_y=y[7:]
plt.scatter(train_x,train_y)
plt.show()
model=np.poly1d(np.polyfit(train_x,train_y,2))
line=np.linspace(0,100,100)
plt.scatter(train_x,train_y)
plt.plot(line,model(line))
plt.show()
```

```python
from sklearn.metrics import r2_score
print(r2_score(train_y,model(train_x)))
```

```python
print(r2_score(test_y,model(test_x)))
```

```python
poly=PolynomialFeatures(degree=2)
x_poly=poly.fit_transform(train_x.reshape(-1,1))

linr=LinearRegression()
linr.fit(x_poly,train_y)
print(linr.predict(poly.fit_transform([[25]])))
```
